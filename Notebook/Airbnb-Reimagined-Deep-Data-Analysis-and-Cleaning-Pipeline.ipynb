### **Data preprocessing**

```python
df.head(2)
```

```python
# Example: Removing outliers from 'minimum nights' using IQR

Q1 = df['minimum nights'].quantile(0.25)
Q3 = df['minimum nights'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['minimum nights'] >= lower_bound) & (df['minimum nights'] <= upper_bound)]
```

```python
# Example: Removing outliers from 'availability 365' using IQR

Q1 = df['availability 365'].quantile(0.25)
Q3 = df['availability 365'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['availability 365'] >= lower_bound) & (df['availability 365'] <= upper_bound)]
```

```python
# Example: Removing outliers from 'calculated host listings count' using IQR

Q1 = df['calculated host listings count'].quantile(0.25)
Q3 = df['calculated host listings count'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['calculated host listings count'] >= lower_bound) & (df['calculated host listings count'] <= upper_bound)]
```

```python
# Example: Removing outliers from 'reviews per month' using IQR

Q1 = df['reviews per month'].quantile(0.25)
Q3 = df['reviews per month'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['reviews per month'] >= lower_bound) & (df['reviews per month'] <= upper_bound)]
```

```python
df.drop(columns=["NAME", "host name"], inplace=True)
```

### **Removed outliers from `minimum nights`, `availability 365`, `calculated host listings count`, and `reviews per month` using the IQR method to improve data quality and model reliability.**

### **Label Encode**

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

categorical_cols = ['host_identity_verified', 'neighbourhood group', 'neighbourhood',
                    'country', 'country code', 'instant_bookable', 'cancellation_policy',"room type"]

for col in categorical_cols:
    df[col] = df[col].astype(str)  
    df[col] = le.fit_transform(df[col])  
```

```python
df = df.drop(columns="service fee")   # Drop it before splitting into X and y

# Then define X and y
X = df.drop(columns="price")
y = df["price"]
```

```python
X.nunique().sort_values(ascending=False)
```

#### **Standard Scaling**

```python
X.head()
```

```python
X['last review'] = pd.to_datetime(X['last review'], errors='coerce')  # handle NaT safely
X['days_since_last_review'] = (pd.Timestamp('today') - X['last review']).dt.days
```

```python
X.drop(columns=["last review","days_since_last_review"], inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

numerical_cols = ["host_identity_verified","neighbourhood group","neighbourhood","lat","long","cancellation_policy","room type","Construction year","minimum nights","number of reviews","reviews per month","review rate number","calculated host listings count","availability 365"]

X[numerical_cols] = scaler.fit_transform(X[numerical_cols]) 
```

```python
X.head()
```

#### **Model Selection**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
linear_model = LinearRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python

linear_model.fit(X_train,y_train)
```

```python
y_pred_linear = linear_model.predict(X_test)
```

```python
print("R2 Score: ",r2_score(y_test,y_pred_linear))
print("MAE: ",mean_absolute_error(y_test,y_pred_linear))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_linear)))
```

```python
y_train_pred = linear_model.predict(X_train)
```

#### **Checking the Overfitting** - **Self doubt becuase of R2 Score**

```python

print("Train R2:", r2_score(y_train, y_train_pred))
print("Test R2:", r2_score(y_test, y_pred_linear))

print("Train RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
print("Test RMSE:", np.sqrt(mean_squared_error(y_test, y_pred_linear)))
```


## ⚠️📌 **Final Note: Target Leakage Detected**

- **During the development of this Airbnb pricing model, the feature service fee was found to have an extremely high correlation (r ≈ 0.998) with the target variable price.**
- **This creates data leakage, where the model learns to predict price using information it shouldn't have in a real-world scenario.**
- **As a result, the Linear Regression model showed an unrealistic R² of 0.99, which collapsed to ~0.0 when service fee was removed.**

---

#### 🛑 **Due to this leakage, the model is invalid for production use.**

- **However, the project remains valuable as a demonstration of:**
- **End-to-end data analysis**
- **Outlier treatment using IQR**
- **Feature encoding and scaling**
- **Model evaluation and overfitting detection**
- **Real-world data issues (like leakage)**

##### ✅ **Lesson: Always check feature-target correlations and guard against data leakage before trusting model metrics.**
